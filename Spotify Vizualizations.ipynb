```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
bpm = pd.read_csv("https://raw.githubusercontent.com/calebs15/spotify/main/return_tables/bpm.csv")
bpm
```

```python
plt.figure(figsize=(20, 6), facecolor = "black")
bars = plt.bar(bpm['bpm'], bpm['avg_streams'], color="limegreen")

# Setting plot background color
plt.gca().set_facecolor('dimgrey')

# Find the bar with the maximum streams
max_streams_bar_index = bpm['avg_streams'].idxmax()
max_streams_bar = bars[max_streams_bar_index]

# Annotate the bar with the maximum streams
plt.annotate("171bpm seems to be a user favorite, with nearly 2.5B average streams",
             xy=(max_streams_bar.get_x() + max_streams_bar.get_width() / 2, max_streams_bar.get_height()),
             xytext=(0, 5),
             textcoords='offset points',
             ha='center',
             color='white')

# Setting tick color
plt.tick_params(axis='x', colors='lightgrey')
plt.tick_params(axis='y', colors='lightgrey')

# Adding labels and title
plt.xlabel('BPM', color='lightgrey', fontsize=14)
plt.ylabel('Average Streams (Billions)', color='lightgrey', fontsize=14)
plt.title('Average Streams per BPM - World\'s Top 950 Most Popular Songs (2024)', color='lightgrey', fontsize=16)




# Show plot
plt.show()
```
